 # matplotlib libeary

```python
import matplotlib.pyplot as plt
```

# Line plot

```python
x1=[1,2,3,4]
y1=[5,6,7,8]
plt.plot(x1,y1,color='red',linestyle="--",linewidth=3)
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

# line plot in different graphs by row and columns

```python
x=[1,2,3,4]
y1=[5,6,7,8]
y2=[9,10,11,12]
plt.subplot(2,1,1)
plt.plot(x,y1)
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.subplot(2,1,2)
plt.plot(x,y2,color="red")
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
```

# bar plot

```python
cri_over=[1,2,3,4,5,6]
cri_run=[8,0,11,24,13,9]
plt.bar(cri_over,cri_run,color="green")
plt.title("Bar plot")
plt.xlabel("over")
plt.ylabel("run")
plt.grid(True)
```

# horizhontal bar plot

```python
cri_over=[1,2,3,4,5,6]
cri_run=[8,0,11,24,13,9]
plt.barh(cri_over,cri_run,color="yellow")
plt.title("Bar plot")
plt.xlabel("over")
plt.ylabel("run")
plt.grid(True)
```

# Scatter plot

```python
x1=[1,2,3,4]
y1=[5,6,7,8]
plt.scatter(x1,y1)
plt.title("Scatter plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

# Pie Chart

```python


# sample data
l = ['A', 'B', 'C', 'D']
x = [15, 30, 45, 10]
c = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

# create a pie chart
plt.pie(x, labels=l, colors=c, autopct='%1.1f%%', startangle=90)

# add a title
plt.title('Pie Chart')

# display the plot
plt.show()
```

# Donut Chart

```python


# sample data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

# create a donut chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, colors=colors, radius=1.5, autopct='%1.1f%%', startangle=90)
ax.pie([100], colors='white', radius=1)

# add a title
ax.set_title('Donut Chart')

# display the plot
plt.show()
```
