```python
import pandas as pd
import plotly.express as px
```

# Load T-Mart data from a CSV file

```python
tmart_data = pd.read_csv('T-Mart.csv')
```

# => Case Study 1: Scatter Plot
## Aesthetics: x-axis represents the DATE, y-axis represents the SALES AMOUNT
## Geometries: Scatter plot to represent each data point
## Scales: Automatically determine the scales for x and y axes
## Coordinates: Cartesian coordinates system

```python
scatter_plot = px.scatter(tmart_data, x='DATE', y='SALES AMOUNT')
```

# Customize scatter plot

```python
scatter_plot.update_layout(
    title='Sales Amount Over Time',
    xaxis_title='Date',
    yaxis_title='Sales Amount'
)
```

## Show the scatter plot

```python
scatter_plot.show()
```

# => Case Study 2: Bar Chart
## Aesthetics: x-axis represents the CATEGORY, y-axis represents the SALES AMOUNT
## Geometries: Vertical bars to represent each category
## Scales: Automatically determine the scales for x and y axes
## Coordinates: Cartesian coordinates system

```python
bar_chart = px.bar(tmart_data, x='CATEGORY', y='SALES AMOUNT')
```

# Customize bar chart

```python
bar_chart.update_layout(
    title='Sales Amount by Category',
    xaxis_title='Category',
    yaxis_title='Sales Amount'
)
```

# Show the bar chart

```python
bar_chart.show()
```

# => Case Study 3: Line Chart
## Aesthetics: x-axis represents the DATE, y-axis represents the SALES AMOUNT
## Geometries: Line connecting the sales amounts for each date
## Scales: Automatically determine the scales for x and y axes
## Coordinates: Cartesian coordinates system

```python
line_chart = px.line(tmart_data, x='DATE', y='SALES AMOUNT')
```

# Customize line chart

```python
line_chart.update_layout(
    title='Sales Amount Trend',
    xaxis_title='Date',
    yaxis_title='Sales Amount'
)
```

# Show the line chart

```python
line_chart.show()
```
